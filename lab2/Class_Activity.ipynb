```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')  # compact, fast, good quality
sentences = [
    "How do I cook pasta?",
    "Best ways to boil spaghetti",
    "What is the capital of France?"
]
embeddings = model.encode(sentences, convert_to_tensor=True)
print("Embeddings shape:", embeddings.shape)

```

```python
from sentence_transformers import util
query = "How to make spaghetti?"
q_emb = model.encode(query, convert_to_tensor=True)
scores = util.pytorch_cos_sim(q_emb, embeddings)[0]
for s, score in zip(sentences, scores):
    print(f"{score.item():.3f}\t{s}")
```

task1 :Basic Pipeline – Sentiment Classification

```python
from transformers import pipeline
import csv

sentiment = pipeline("sentiment-analysis")

# Create your own dataset (example: 200 simple sentences)
texts = [f"This is sample sentence number {i}. I like it." for i in range(200)]

results = sentiment(texts)

# Save to CSV
with open("task1_sentiment_200.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["text", "label", "score"])
    for t, r in zip(texts, results):
        writer.writerow([t, r['label'], r['score']])

print("Task 1 completed: CSV saved")
```

Task 2: Summarization of News Articles (15 marks)

```python
from transformers import pipeline

summarizer = pipeline("summarization")

articles = [
    "Artificial intelligence is transforming healthcare by improving diagnosis and treatment.",
    "Climate change is affecting weather patterns across the globe.",
    "The stock market saw significant growth today due to tech stocks.",
    "Education systems are adapting to online learning platforms.",
    "Renewable energy sources are gaining popularity worldwide.",
    "Space exploration has advanced with new satellite launches.",
    "Cybersecurity threats are increasing with digital transformation.",
    "Electric vehicles are becoming more affordable.",
    "Machine learning is widely used in recommendation systems.",
    "Global economies are recovering after the pandemic."
]

for i, art in enumerate(articles, 1):
    summary = summarizer(art, max_length=40, min_length=15, do_sample=False)
    print(f"\nArticle {i} Summary:")
    print(summary[0]['summary_text'])
```

Translation – English → French → English

```python
from transformers import pipeline

en_fr = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
fr_en = pipeline("translation_fr_to_en", model="Helsinki-NLP/opus-mt-fr-en")

sentences = [
    "Machine learning is very powerful.",
    "The weather is nice today.",
    "I enjoy studying computer science.",
    "This lab is interesting.",
    "Artificial intelligence is the future.",
    "Data science combines statistics and programming.",
    "FastAI simplifies deep learning.",
    "Transformers are widely used in NLP.",
    "Python is easy to learn.",
    "Cloud computing is scalable."
]

for s in sentences:
    fr = en_fr(s)[0]['translation_text']
    back = fr_en(fr)[0]['translation_text']
    print("\nOriginal:", s)
    print("French:", fr)
    print("Back to English:", back)
```

4. Embeddings & similarity: Build embeddings for 100 FAQs and implement a simple semantic search (query -> top 3 answers)

> Add blockquote



```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

# Example FAQs (extend to 100 similarly)
faqs = [f"How do I reset my password? Step {i}" for i in range(100)]
embeddings = model.encode(faqs, convert_to_tensor=True)

query = "How can I change my password?"
q_emb = model.encode(query, convert_to_tensor=True)

scores = util.pytorch_cos_sim(q_emb, embeddings)[0]
top_results = scores.topk(3)

print("Top 3 similar FAQs:")
for idx in top_results.indices:
    print("-", faqs[idx])
```
