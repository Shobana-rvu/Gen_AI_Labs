```python
from fastai.vision.all import *
from duckduckgo_search import DDGS
import time, random

def search_images(term, max_images=20, max_retries=5):
    print(f"Searching for '{term}'...")

    for attempt in range(max_retries):
        try:
            with DDGS() as ddgs:
                results = ddgs.images(
                    keywords=term,
                    max_results=max_images,
                    safesearch="Moderate",
                    region="in-en"
                )
                urls = [r["image"] for r in results]

            # random wait to avoid ban
            time.sleep(random.uniform(2, 5))
            return L(urls)

        except Exception as e:
            print(f"Attempt {attempt+1}/{max_retries} failed due to: {e}")
            time.sleep(8 + attempt*4)

    return L([])
```

```python
searches = 'forest','bird'
path = Path('bird_or_not')

for o in searches:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)

    urls = search_images(f'{o} photo', max_images=8)

    print(f"Downloading images for: {o}")
    download_images(dest, urls=urls)

print("Download done!")
```

```python
failed = verify_images(get_image_files(path))
failed.map(Path.unlink)
print("Removed corrupted images:", len(failed))
```

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(path, bs=16, num_workers=0)

from fastai.callback.progress import ProgressCallback
from fastprogress.fastprogress import force_console_behavior

force_console_behavior()

learn = vision_learner(dls, resnet18, metrics=accuracy)

# remove notebook progress callback
learn.remove_cbs(ProgressCallback)

learn.fine_tune(1)
```

```python
import requests

def download_image(url, filename="test_bird.jpg"):
    headers = {"User-Agent": "Mozilla/5.0"}
    r = requests.get(url, headers=headers, timeout=20)
    r.raise_for_status()
    with open(filename, "wb") as f:
        f.write(r.content)
    return filename
```

```python
urls = search_images("bird photo", max_images=5)

downloaded = False
for u in urls:
    try:
        file = download_image(u, "test_bird.jpg")
        downloaded = True
        print("Downloaded from:", u)
        break
    except Exception as e:
        print("Failed URL:", u, "|", e)

if downloaded:
    pred, pred_idx, probs = learn.predict(PILImage.create("test_bird.jpg"))
    print(f"This is a: {pred}.")
    print(f"Probability it's a bird: {probs[pred_idx]:.4f}")
else:
    print("All URLs blocked. Try rerun search or use a different query like 'bird wildlife photo'.")
```
